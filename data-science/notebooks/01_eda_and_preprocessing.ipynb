## 4. Preprocessing & Text Cleaning
Fungsi pembersihan teks untuk membersihkan karakter khusus, email, link website, dan spasi berlebih pada teks CV mentah.

```python
import re

def clean_resume_text(text):
    if not isinstance(text, str):
        return ""
    
    # Ubah menjadi lowercase
    text = text.lower()
    
    # Hapus URL/Hyperlink
    text = re.sub(r'http\S+\s*', ' ', text)
    
    # Hapus RT dan cc (jika ada text sisaan social media)
    text = re.sub(r'rt|cc', ' ', text)
    
    # Hapus hashtags
    text = re.sub(r'#\S+', ' ', text)
    
    # Hapus mention (@username)
    text = re.sub(r'@\S+', '  ', text)
    
    # Hapus email
    text = re.sub(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b', ' ', text)
    
    # Hapus karakter khusus dan angka jika dirasa tidak perlu (atau sesuaikan dengan kebutuhan)
    text = re.sub(r'[^a-z\s]', ' ', text)
    
    # Hapus spasi berlebih
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

# Contoh penggunaan
sample_text = "CV John Doe - Email: john.doe@email.com, Web: https://johndoe.dev. Skills: Python, SQL, Machine Learning!"
print("Sebelum:", sample_text)
print("Sesudah:", clean_resume_text(sample_text))
```
